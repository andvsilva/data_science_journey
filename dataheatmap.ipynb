```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Seu DataFrame com índices Q1...Q5
df = pd.DataFrame({
    '1A': ['51%', '62%', '35%', '8%', '5%'],
    '1B': ['80%', '66%', '43%', '34%', '20%'],
    '1C': ['71%', '54%', '34%', '26%', '69%'],
    '1D': ['63%', '53%', '63%', '16%', '59%'],
    '1E': ['68%', '68%', '49%', '16%', '59%'],
    '1F': ['67%', '64%', '45%', '21%', '24%'],
    '1G': ['65%', '50%', '38%', '38%', '50%'],
}, index=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'])

# Converter de string '51%' para float 51
df_numeric = df.applymap(lambda x: float(x.strip('%')))

# Plotar o heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_numeric, annot=True, fmt=".0f", cmap="coolwarm")
plt.title("Heatmap de Percentuais (%)")
plt.show()
```
