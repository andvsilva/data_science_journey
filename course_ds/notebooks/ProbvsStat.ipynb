```python
def f_getPowerSet(S):
    pS = set()
    A = np.array(list(S))
    n = A.size
    for i in range(2**n):
        b = np.binary_repr(i, width=n)
        idx = np.array(list(b),dtype=int)==1
        pS.add(frozenset(A[idx]))
    return pS
```

```python
S = set(np.arange(3))
pS = f_getPowerSet(S)
```

```python
for i in pS:
    print(i)
```

```python
def BernoulliTrail(p=0.5):
    X = int(np.random.rand() <= p)
    return X
```

```python
print(BernoulliTrail())
```

```python
countOnes = 0
countZeros = 0
p = 0.611334
n = 10000
for i in range(n):
    x = BernoulliTrail(p)
    if x == 1:
        countOnes +=1
    else:
        countZeros +=1
print(f"Success fraction (1) is {countOnes/n}")
print(f"Success fraction (0) is {countZeros/n}")
```

```python
def GT(p):
    X = 1
    while True:
        if BernoulliTrail(p) == 1:
            return X
        else:
            X+=1
```

```python
p = 0.7
print(GT(p))
```

### Expectation - Law of Large Numbers

#### Bernoulli
 
$E[X] = p$

```python

 
p = 0.812
sMeanX = []
for N in range(1,100000,100):
     sMeanX.append((np.random.rand(N)<=p).mean())

y = np.array(sMeanX)
x = np.arange(y.size)

plt.plot(x,y,linestyle='--')
 
```

## Geometric(p), $[X] = 1/p$

```python
# E[X] = 1/p
 
p = 0.312
print(1/p)
sMeanX = []
for N in range(1,100000,100):
     sMeanX.append(np.random.geometric(p,N).mean())

y = np.array(sMeanX)
x = np.arange(y.size)

plt.plot(x,y,linestyle='--')
```

### Binomial(n,p) $E[x] = np$

```python
# E[X] = np
n = 350
p = 0.312
print(f"E[x] = {n*p}")
sMeanX = []
for N in range(1,100000,100):
     sMeanX.append(np.random.binomial(n,p,N).mean())

y = np.array(sMeanX)
x = np.arange(y.size)

plt.plot(x,y,linestyle='--')
```

## Normal($\mu$,$\sigma$),  $E[X] = \mu$

```python
# E[X] = mu
mu = 2.4
sgma = 3.1

print(f"E[x] = {mu}")
sMeanX = []
for N in range(1,100000,100):
     sMeanX.append(np.random.normal(mu,sgma,N).mean())

y = np.array(sMeanX)
x = np.arange(y.size)

plt.plot(x,y,linestyle='--')
```
